```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statistics
```

```python
# Load csv files
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
# City data into Dataframe
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head()
```

```python
# Ride data into Dataframe
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head()
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data_df.head()
```

```python
# Create the Urban city DataFrame.
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
urban_cities_df.head()
```

```python
# Create the Suburban city DataFrames.
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
suburban_cities_df.head()
```

```python
# Create the Rural city DataFrames.
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
rural_cities_df.head()
```

```python
# Get the total number of rides for each city type.
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# Get the total number of drivers for each city type.
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
# Calculate the sum of fares by each city type.
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
# Calculate the average fare per ride.
average_fare_per_ride = total_fares / total_rides
average_fare_per_ride
```

```python
# Calculate the average fare per driver.
average_fare_per_driver = total_fares / total_drivers
average_fare_per_driver
```

```python
# Create a list of strings 
lst = [total_rides, total_drivers, total_fares, average_fare_per_ride,  
            average_fare_per_driver]
summary = pd.DataFrame(lst)
summary
```

```python
# Change the names of indexes 
summary.index = ['Total Rides', 'Total Drivers', 'Total Fares', 'Average Fare per Ride', 'Average Fare per Driver'] 
summary
```

```python
# Switch index with columns and delete the index name.
summary_df = summary.swapaxes("index", "columns")
summary_df.index.name = None
summary_df
```

```python
# Format each column
summary_df['Total Rides'] = summary_df['Total Rides'].map('{:,.0f}'.format)
summary_df['Total Drivers'] = summary_df['Total Drivers'].map('{:,.0f}'.format)
summary_df['Total Fares'] = summary_df['Total Fares'].map('${:,.2f}'.format)
summary_df['Average Fare per Ride'] = summary_df['Average Fare per Ride'].map('${:,.2f}'.format)
summary_df['Average Fare per Driver'] = summary_df['Average Fare per Driver'].map('${:,.2f}'.format)
summary_df
```

```python
# Raname columns for the merged dataframe.
pyber_data_df.columns = ['City', 'Date', 'Fare', 'Ride Id', 'No. Drivers', 'City Type']
pyber_data_df.head()
```

```python
# Set index to the Date column.
pyber_data_df.set_index('Date').head()
```

```python
# Create new dataframe including only the Date, City Type, and Fare columns.
pyber_fare = pyber_data_df[['Date', 'City Type', 'Fare']].copy()
pyber_fare_df = pyber_fare.set_index("Date")
pyber_fare_df.head()
```

```python
# Set the index to the datetime data type.
pyber_fare_df.index = pd.to_datetime(pyber_fare_df.index)
pyber_fare_df.info()
```

```python
# Get the total number of drivers for each city type.
pyber_total_fare = pyber_fare_df.groupby(["City Type"]).sum()["Fare"]
pyber_total_fare
```

```python
# Reset index.
pyber_fare_df.reset_index().head()
```

```python
# Create pivot table dataframe 
pyber_table = pd.pivot_table(pyber_fare_df, values ='Fare', index =['Date'], 
                         columns =['City Type'], aggfunc = np.sum)
pyber_table.head()
```

```python
# New dataframe with given dates
pyber_specific = pyber_table.loc['2019-01-01':'2019-04-28']
pyber_specific.head()
```

```python
# New dataframe in weekly bins & sum of the fares of each week.
pyber_fare_week = pyber_specific.resample('W').sum()
pyber_fare_week
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')
pyber_fare_week.plot(figsize=(10,5));
plt.suptitle("Total Fare by City Type")
plt.xlabel('Month', fontsize=18)
plt.ylabel('Fare($USD)', fontsize=16)
# Save the figure.
plt.savefig("analysis/PyBer_Challenge figure.png")
```
